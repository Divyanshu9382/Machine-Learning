# **Feature Selection**

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula="Sales ~ TV ", data=df).fit()
lm.summary()
```

```python
lm = smf.ols(formula="Sales ~ TV+Radio ", data=df).fit()
lm.summary()
```
